```python
import pandas as pd
from sklearn.model_selection import train_test_split
from pytorch_tabular import TabularModel
from pytorch_tabular.models import NodeConfig
from pytorch_tabular.config import DataConfig, TrainerConfig, OptimizerConfig
```

```python
df = pd.read_csv(r"C:\Users\dbastola2022\OneDrive - Florida Atlantic University\Academics\Research\Malnutrition\MICS\malnutrition\Dataset\ch.csv") #Local
df.head(2)
```

```python
from pytorch_tabular.config import OptimizerConfig

optimizer_config = OptimizerConfig()
```

```python
from pytorch_tabular import TabularModel
from pytorch_tabular.models.node.config import NodeConfig
from pytorch_tabular.config import DataConfig, TrainerConfig, OptimizerConfig

# Define columns
target_col = "malnurished"
continuous_cols = ["child_weight"]
categorical_cols = [col for col in df.columns if col not in continuous_cols + [target_col]]

# Split dataset
train, test = train_test_split(df, test_size=0.2, stratify=df[target_col], random_state=42)

# Configurations
data_config = DataConfig(
    target=[target_col],  # must be a list
    continuous_cols=continuous_cols,
    categorical_cols=categorical_cols,
)

model_config = NodeConfig(
    task="classification",
    num_layers=2,
    num_trees=256,
    depth=6,
    learning_rate=1e-2,
    metrics=["accuracy", "precision", "recall"],
)

trainer_config = TrainerConfig(
    auto_lr_find=True,
    batch_size=256,
    max_epochs=100,
    accelerator="cpu"
)

optimizer_config = OptimizerConfig() 

# Build and train model
tabular_model = TabularModel(
    data_config=data_config,
    model_config=model_config,
    trainer_config=trainer_config,
    optimizer_config=optimizer_config,
)

tabular_model.fit(train=train, validation=test)

# Evaluate
results = tabular_model.evaluate(test)
print(results)
```

```python
from sklearn.metrics import f1_score

preds = tabular_model.predict(test)
f1 = f1_score(test["malnurished"], preds["prediction"])
print(f"F1 Score: {f1:.4f}")
```
