```python
# Calculate missing values
missing_values = df1.isna().sum()

# Create a list of dictionaries for each column
data = []
for var_name, var_label in zip(meta.column_names, meta.column_labels):
    data.append({
        "Variable Name": var_name,
        "Variable": var_label,
        "missing values": missing_values.get(var_name, 0)
    })

# Create a DataFrame
missing_df = pd.DataFrame(data)

# Save to Excel
missing_df.to_excel("fs.xlsx", index=False)

print("Excel file 'fs.xlsx' created.")
```
